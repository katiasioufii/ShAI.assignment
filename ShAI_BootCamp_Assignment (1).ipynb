```python
print('mean - median ')
print(f"mean :{df['TotalPay'].mean()}")
print(f" median : {df['TotalPay'].median()}")
print( f"mode : {df['TotalPay'].mode()}")
print(f"minimum :{df['TotalPay'].min()}")
print(f"maximum : {df['TotalPay'].max()}")
print(f"range of the salary : {df['TotalPay'].max() - df['TotalPay'].min()}")
print( f"stander diviation : {df['TotalPay'].std()}")
```

```python
avg_base_pay = df['BasePay'].mean()
df['BasePay'] = df['BasePay'].fillna(avg_base_pay)

# Fill missing values in 'OvertimePay' column with the average of non-missing values
avg_overtime_pay = df['OvertimePay'].mean()
df['OvertimePay'] = df['OvertimePay'].fillna(avg_overtime_pay)


df['Benefits'].fillna(df['TotalPayBenefits']- df['OvertimePay'], inplace=True)
#because there is no values i can't feel them myself  i will drop them
df.drop('Notes', axis=1, inplace=True)
df.drop('Status', axis=1, inplace=True)
```

```python
for column in df.columns:
    column_data_type = df[column].dtype
    missing_value_count = df[column].isnull().sum()
    print(f'Column: {column}')
    print(f'Data Type: {column_data_type}')
    print(f'Missing Value Count: {missing_value_count}\n')
```

```python
df['JobTitle1'] = df['JobTitle'].str.replace(r'\(|\)', '')
df['JobTitle1'] = df['JobTitle'].str.split(',').str[-1]
#print(df['JobTitle1'])
# Extract the department from the modified job title
df['Department'] = df['JobTitle1'].str.split('(').str[-1]
df['Department'] = df['Department'].str.replace(')', '')
#df['Department'] = df['JobTitle1'].str.split('(').str[-1].str.rstrip(')')

# Check if the department is split into one or two entries
#df['Department'] = df['JobTitle'].apply(lambda x: x if (',' in x) or ('(' in x) else 'No Department')

# Print the updated DataFrame
print(df['Department'] .head(10))
```

```python
department_mapping = {
     'bureau of fire prevention and public s': 'bureau of fire prevention and public safty',
    'bureau of fire prevention and public sa': 'bureau of fire prevention and public safty',
       'bureau of fire prevention and public saf': 'bureau of fire prevention and public safty',
       'bureau of fire prevention and public saft': 'bureau of fire prevention and public safty',
    'fire suppress' : 'fire suppression',
    'fire suppressi' : 'fire suppression',

}

# Clean the department names
df['Department'] = df['Department'].map(department_mapping).fillna(df['Department'])

# Print the updated DataFrame
print(df['Department'])
```

```python
df['Department'] = df['Department'].str.lower()
df['Department'].unique()
```

```python
def clean_department_name(department):
    department = department.strip()  # Remove extra spaces around the word
    for substring in df['Department'].unique():
        if substring in department and substring != department:
            return substring
    return department

df['Department'] = df['Department'].apply(clean_department_name)
```
